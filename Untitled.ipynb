# ** Uber Data Analysis with Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import calendar
```

```python
data = pd.read_csv('Uber Drives - .csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.isnull().any()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['START_DATE*'] = pd.to_datetime(data['START_DATE*'],format = '%m/%d/%Y %H:%M')
data['END_DATE*'] = pd.to_datetime(data['END_DATE*'],format = '%m/%d/%Y %H:%M')
```

```python
data.dtypes
```

```python
hour = []
day = []
dayofweek = []
month = []
weekday = []

for x in data['START_DATE*']:
    hour.append(x.hour)
    day.append(x.day)
    dayofweek.append(x.dayofweek)
    month.append(x.month)
    weekday.append(calendar.day_name[dayofweek[-1]])
    
data['HOUR'] = hour
data['DAY'] = day
data['DAY OF WEEK'] = dayofweek
data['MONTH'] = month
data['WEEKDAY'] = weekday
```

```python
data.head()
```

```python
data['CATEGORY*'].value_counts() #Categories we have
```

```python
sns.countplot(x = 'CATEGORY*',data = data)
```

```python
# How long do people travel with Uber
data['MILES*'].plot.hist()
```

```python
# What hour do most people take uber to their destination
hours = data['START_DATE*'].dt.hour.value_counts()

hours.plot(kind = 'bar',color = 'red', figsize=(10,5))

plt.xlabel('Hours')
plt.ylabel('Frequency')

plt.title('Number of trips and hours')

plt.show()
```

```python
# Check the purpose of trips
data['PURPOSE*'].value_counts().plot(kind = 'bar', figsize=(10,5),color = 'brown')
```

```python
# Which Day has the highest number of trips

data['WEEKDAY'].value_counts().plot(kind = 'bar',color = 'blue', figsize = (10,5))
plt.show()
```

```python
data['DAY'].value_counts().plot(kind = 'bar',color = 'red',figsize=(10,5))
plt.show()
```

```python
# What are the trips in the month
data['MONTH'].value_counts().plot(kind = 'bar',color = 'green',figsize=(10,5))
plt.show()
```

```python
#Where do people boarding their trip from most?
data['START*'].value_counts().plot(kind = 'bar',color = 'orange',figsize=(25,10))
```
